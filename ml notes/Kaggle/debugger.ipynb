```python
import os

def print_tree(startpath, indent="  "):
    """
    Prints a tree structure of all files and folders in a given directory.

    Parameters:
    - startpath (str): The root directory to generate the tree from.
    - indent (str): The characters used for indentation.
    """
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent_space = indent * level
        print(f"{indent_space}üìÅ {os.path.basename(root)}/")
        subindent_space = indent * (level + 1)
        for file in files:
            print(f"{subindent_space}üìÑ {file}")

# Replace with your target directory
path = r"solution-lmsys-chatbot-arena"
print("Directory Tree:")
print_tree(path)
```
